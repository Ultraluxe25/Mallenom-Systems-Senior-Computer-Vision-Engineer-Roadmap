# Декоратор property

Декоратор property дает доступ к атрибутам класса в методе. Название этого метода с частичкой `.setter` делает его сеттером и также геттером и делиттером

```Python
@property
def <название метода-интерфейса>:
    ...
    доступ к атрибутам
    ...
    return <атрибуты>

@<название метода-интерфейса>.getter
def <геттер>
    ...

@<название метода-интерфейса>.setter
def <сеттер>
    ...

@<название метода-интерфейса>.deleter
def <делитер>
    ...
```

Моя попытка разобраться с сеттерами и геттерами (мини проект): 

```python
class Diary:
    def __init__(self, secrets: list[str]):
        self.__secrets = secrets

    @property
    def say_secrets(self) -> str:
        for secret in self.__secrets:
            print(secret)

    @say_secrets.getter
    def say_rumor(self) -> list[str]:
        return self.__secrets

    @say_secrets.setter
    def add_secret(self, value):
        self.__secrets.append(value)

    @say_secrets.deleter
    def lost_diary_key(self):
        del self.__secrets
```
