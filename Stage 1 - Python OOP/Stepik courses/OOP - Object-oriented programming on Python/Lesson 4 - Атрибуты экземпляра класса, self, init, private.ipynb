Сначала ищем атрибут в локальном пространстве имен, а если нету, тогда в пространстве имен класса

Для экземпляра класса функция является методом

### Декоратор @staticmethod мы можем вызывать для вызова из экземпляра класса методов, которые не имеют self в объявлении

# Self

Методы производит действие над объектом, который его вызывает (инициализированны внутри класса)

**self** - это аргумент на место которого перемещается объект класса

# Init

Магический метод отвечает за те действия и создание атрибутов, которые произойдут в каждый раз создания нового экземпляра класса

# Моносостояние



При изменении состояния (атрибута) одного экземпляра, так же изменятся все экземпляры одновременно (могут быть созданы новые поля экзепляров)

Для хранения нужно использовать словарь, т.к. это изменяемый объект и все экземпляры будут обращаться к единому объекту

# Public, protected, private или введение в инкапсуляцию

**protected** на уровне согласования означает что данный атрибут (метод, поле) лучше вызывать из класса, а не через объект

**private** позволяет легко обращаться к таким переменным внутри реализации класса, но снаружи его нельзя использовать! (можно через <название экземпляра>._<название класса>__<приватный атрибут>)



В данном случае мы получили доступ именно через метод класса, но не напрямую. Напрямую к таким атрибутам доступ ограничен



### Для настоящей инкапсуляции нужно использовать модуль **accessify** с декораторами _@protected_ и _@private_
