По приоритету моржовая операция отрабатывает в последнюю очередь



Моржовый операция позволяет избежать двух импутов



Та же логика что и в примере выше



Иногда программа работает неопределенным способом и может привести к ошибки



Из-за низкой приоритетности нужно ставить скобки вокруг моржового оператора

### Моржовый оператор нужно использовать лишь в целях улучшения производительности и читабельности программ
