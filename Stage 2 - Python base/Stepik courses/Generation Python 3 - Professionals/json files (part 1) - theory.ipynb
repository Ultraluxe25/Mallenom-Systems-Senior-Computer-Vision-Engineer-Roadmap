# Тема урока: работа с json файлами

**JSON** (англ. JavaScript Object Notation) — текстовый формат обмена данными, основанный на языке программирования JavaScript.

**Сериализация** — это процесс преобразования объекта (например, структуры данных, экземпляра класса) в формат json.

#### В отличие от словаря, ключи в json могут быть только строками, заключенными в двойные кавычки.

Преобразование переменных программы (Python-объектов) в формат для хранения называется сериализацией, а обратное преобразование — десериализацией. В Python для сериализации и десериализации в формат json есть модуль, который так и называется — json.
