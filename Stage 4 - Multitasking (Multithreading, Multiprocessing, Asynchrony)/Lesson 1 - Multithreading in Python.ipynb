**Многозадачность** - способность ОС выполнять несколько задач одновременно.

Процессная многозадачность основана на процессах — _одновременно выполняющихся_ программах. Процесс — это _экземпляр программы_, который выполняется в оперативной памяти компьютера и обладает _собственным адресным пространством_, в котором хранятся данные и переменные, необходимые для выполнения программы. Каждый процесс имеет _уникальный идентификатор (PID)_, позволяющий _операционной системе управлять его выполнением и ресурсами_. Ключевая особенность процессной многозадачности заключается в том, что _процессы полностью изолированы друг от друга — они не имеют прямого доступа к памяти и ресурсам других процессов без специальных механизмов межпроцессного взаимодействия (IPC)_. Даже процессы, принадлежащие одному пользователю, операционная система рассматривает как независимые сущности. Прямо сейчас откройте Диспетчер задач (Windows) -> первая вкладка Процессы или системный монитор (Linux) и посмотрите, как много процессов выполняется одновременно.
